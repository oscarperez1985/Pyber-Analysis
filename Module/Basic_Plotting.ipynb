## Input Data

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

### Create a Line Graph

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() (i.e. the object-oriented interface method)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# The above graph can also be plotted in this way: Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
##### This is how the plot is tipycally plotted:
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
##### Add axis lables and title/legend
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
##### Format the plot:
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create and format the graph using the object-oriented method:
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color='green', marker='D', linewidth=2, label="Boston")
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.grid()
```

## Scatter Plot (Object Oriented-Method)

```python
### Repeat input data for reference
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

y_axis5 = []
for i in y_axis:
    scaled = i * 5
    y_axis5.append(scaled)
y_axis5
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=y_axis5, label='Boston', c='skyblue', edgecolor='black', alpha = 0.8)
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
## Invert the y-axis
plt.gca().invert_yaxis()
ax.set_xlim(0,50)
```
